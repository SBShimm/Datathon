```python
import pandas as pd
import numpy as np
```

```python
netflix = pd.read_csv('./netflix_titles.csv')
imdb = pd.read_csv('./netflix_imdb.csv')
```

```python
netflix.head()
```

```python
imdb_join = imdb[['Title', 'Imdb Score']]
imdb_join.columns = ['title', 'Imdb']
```

```python
imdb_join['IMDB'] = imdb_join.Imdb.str.extract(r'(\d+.\d+)')
del imdb_join['Imdb']
```

```python
imdb_join.head()
```

```python
data_join = pd.merge(left=netflix, right=imdb_join, how='left', on='title')
```

```python
data_join.head()
```

```python
data_join['IMDB'].isnull().sum()
```

```python
data_join['first_country'] = data_join.country.str.extract(r'([a-zA-Z\s]+)')
```

```python
data_join
```

```python
data_join.to_csv('Netflix_SBShim_07261.csv')
```
